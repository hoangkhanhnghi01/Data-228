```python
combined = pd.read_csv("/content/drive/MyDrive/Data228/Cleaned Data/Normalized combined table/new_combined_table.csv")
df = pd.DataFrame(combined)
```

```python
df.isna().sum()
```

```python
df["age_0_17_deaths"].ffill(inplace=True)
df["age_18_49_deaths"].ffill(inplace=True)
df["age_50_64_deaths"].ffill(inplace=True)
df["age_65_deaths"].ffill(inplace=True)
df["gender_female_deaths"].ffill(inplace=True)
df["gender_male_deaths"].ffill(inplace=True)
df["race_white_deaths"].ffill(inplace=True)
df["race_other_deaths"].ffill(inplace=True)
df["race_multirace_deaths"].ffill(inplace=True)
df["race_latino_deaths"].ffill(inplace=True)
df["race_hawaiian_deaths"].ffill(inplace=True)
df["race_black_deaths"].ffill(inplace=True)
df["race_asian_deaths"].ffill(inplace=True)
df["race_american_deaths"].ffill(inplace=True)
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
#df = df[(df['date'] > '2020-04-21') & (df['date'] < '2022-02-22')]
```

```python
df.isna().sum()
```

```python
x = df.iloc[:,1:].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
```

```python
normalized = pd.DataFrame(df["date"]).join(pd.DataFrame(x_scaled))
```

```python
normalized.columns = list(df)
```
